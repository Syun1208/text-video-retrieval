```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
import json
import glob
path = "/content/drive/MyDrive/AI_Challenge/ASR/text"
path_json = "/content/drive/MyDrive/AI_Challenge/ASR/metadata2"
ls_json = sorted(glob.glob(f"{path_json}/*"))

with open(f"{path}/text2.txt", "w") as file:
    i =0

    for js in ls_json:

      with open(js) as json_file:
          data = json.load(json_file)
      url = data["watch_url"]
      video_id = url.split("v=")[-1]

      try:
        i +=1
        print(i)
        script = YouTubeTranscriptApi.get_transcripts([video_id], languages=['de', 'vi'])
        flag = "yes"
      except:
        flag = "no"

      if flag == "yes":
        dd = len(script[0][video_id])
        for i in range(dd):

            print("a")
            start = script[0][video_id][i]["start"]
            duration = script[0][video_id][i]["duration"]
            end = start + duration
            tranfor_start = int(start * 25)
            trafor_end = int(end * 25)

            name_L = js.split("/")[-1].split(".")[0]
            print(name_L)
            file.write(f"{name_L},{tranfor_start},{trafor_end}," + script[0][video_id][i]["text"] + "\n")
```

```python
def merge_files(file1, file2, file3, output_file):
    with open(output_file, "w") as output:
        with open(file1, "r") as f1:
            content1 = f1.read()
            output.write(content1)

        with open(file2, "r") as f2:
            content2 = f2.read()
            output.write(content2)

        with open(file3, "r") as f3:
            content3 = f3.read()
            output.write(content3)

    print("Đã gộp thành công các file thành một file mới.")

# Đường dẫn đến các file cần gộp
file1 = "/content/drive/MyDrive/AI_Challenge/ASR/text/text1.txt"
file2 = "/content/drive/MyDrive/AI_Challenge/ASR/text/text2.txt"
file3 = "/content/drive/MyDrive/AI_Challenge/ASR/text/text3.txt"

# Đường dẫn đến file kết quả gộp
output_file = "/content/drive/MyDrive/AI_Challenge/ASR/text/merge_files.txt"

# Gọi hàm gộp file
merge_files(file1, file2, file3, output_file)
```
